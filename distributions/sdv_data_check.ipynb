```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
features_to_plot = ['People_Killed', 'Min_Civilians', 'Max_Civilians']
real_df = pd.read_csv("../data/final_cleaned_drone_dataa.csv")
synthetic_df = pd.read_csv("../data/synthetic_data_ctgan_full.csv")
```

```python
# Plotting
for feature in features_to_plot:
    plt.figure(figsize=(8, 4))
    
    sns.histplot(real_df[feature], kde=True, color='blue', label='Real', stat='density', bins=30, alpha=0.6)
    sns.histplot(synthetic_df[feature], kde=True, color='orange', label='Synthetic', stat='density', bins=30, alpha=0.6)
    
    plt.title(f'Distribution Comparison: {feature}')
    plt.xlabel(feature)
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
